##Pre-processing of the datafiles
Open the text files

```python
short_pos = open("text_data/Positive.txt","r").read()
short_neg = open("text_data/Negative.txt","r").read()
```


Split the sentence into individual words, also called word tokenization.

```python
short_pos_words = word_tokenize(short_pos)
short_neg_words = word_tokenize(short_neg)
```

print short_pos_words

```python
print(short_pos_words)
```

Import stopwords for experimental purposes

```python
nltk.download("popular")
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
```

Check importance of stops

```python
print(stops)
```

Store all punctuations in a string

```python
import string
all_punctuation = string.punctuation
```

Filter the pos and neg words.

Remove all punctuations from short_pos_words & short_neg_words 
Finally store all the short_pos_words & short_neg_words in lower case

```python
#removales used for experimental pre-processing
removales = list(stops)+list(all_punctuation)+list("n't")
filteredPosWords = []

for w in short_pos_words:
    w_lower = w.lower()
    if  (w_lower not in all_punctuation):
        filteredPosWords.append(w_lower)

print(filteredPosWords)
```

```python
#removales used for experimental pre-processing
removales = list(stops)+list(all_punctuation)+list("n't")
filteredNegWords = []

for w in short_neg_words:
    w_lower = w.lower()
    if(w_lower not in all_punctuation):
        filteredNegWords.append(w_lower)

print(filteredNegWords)
```

##Create a list called all_words that will contain all the filtered words in the dataset 

```python
all_words = []
```

```python
#Add all positive and negative words to the list
for w in filteredPosWords:
    all_words.append(w)

for w in filteredNegWords:
    all_words.append(w)
```

```python
print(all_words)
```


Store the frequency of each word in all_words by using 'nltk.FreqDist(all_words)'

```python
all_words = nltk.FreqDist(all_words)
```

Check 20 most common words

```python
print(all_words.most_common(20))
```

Print the total number of all words

```python
print(all_words)
```

##Create word_features and do feature extraction

Use the first 500 keys from all_words as word_Features

```python
word_features = list(all_words.keys())[:500]
```

Create a document that holds all the positive and negative sentences in the text files.

```python
documents = []
```

```python
for r in short_pos.split('\n'):
    documents.append( (r, "pos") )

for r in short_neg.split('\n'):
    documents.append( (r, "neg") )
```

Extract features from the documents

```python
featuresets = extract_features(documents, word_features)
```

##Form training and test datasets

Shuffle featuresets before creating training and test datasets

```python
random.shuffle(featuresets)
```

We allocate the first 150 sentences for training the dataset and the last 51 for testing

```python
training_set = featuresets[:151]
testing_set =  featuresets[151:]
```

## Applying Naive Bayes classifier
* Train using nltk
* Test using the last 51 sentences in featuresets called testing_set

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

Print accuracy percentage


```python
print("Original Naive Bayes Algo accuracy percent:", (nltk.classify.accuracy(classifier, testing_set))*100)
```
